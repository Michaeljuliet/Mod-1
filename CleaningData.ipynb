```python
# import the library to be used 
import pandas as pd
import numpy as np
```

```python

# Load the dataset from a text file
file_path = 'C://Users//USER//Documents//dataset.txt'
df = pd.read_csv(file_path)

# Display the DataFrame to verify its contents
df.head(10)
```

```python
df.isnull().sum()
```

```python
# Handle missing values for Age by imputing with the median
df['Age'].fillna(df['Age'].median(), inplace=True)

# Handle missing Price by imputing with the mean
df['Price'].fillna(df['Price'].mean(), inplace=True)

# Handle missing RatingOfProduct by imputing with the mode
df['RatingOfProduct'].fillna(df['RatingOfProduct'].mode()[0], inplace=True)

# Handle missing AdvertisingAgency by imputing with the mode
df['AdvertisingAgency'].fillna(df['AdvertisingAgency'].mode()[0], inplace=True)

# Handle missing custName and Product with "Unknown"
df['custName'].fillna("Unknown", inplace=True)
df['Product'].fillna("Unknown", inplace=True)

# Display the cleaned DataFrame to verify the changes
df.head()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
# Check for outliers in Price using the IQR method
Q1 = df['Price'].quantile(0.25)
Q3 = df['Price'].quantile(0.75)
IQR = Q3 - Q1

# Define a range to filter out outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter the DataFrame to remove outliers in Price
df = df[(df['Price'] >= lower_bound) & (df['Price'] <= upper_bound)]

# Display the cleaned DataFrame without outliers
df.head()
```

```python
# Save the cleaned DataFrame to an Excel file
df.to_excel('cleaned_customer_data.xlsx', index=False)

```
